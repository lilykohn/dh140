```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.gutenberg.org/files/23042/23042-h/23042-h.htm'
response = requests.get(url)
document = response.text
```

```python
document
```

```python
document = BeautifulSoup(response.text, "html.parser")
```

```python
temptext = document.find('body').text 
```

```python
print(temptext)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
sentences = sent_tokenize(temptext)
```

```python
sentences
```

```python
#Tokenize the words and remove stopwords
```

```python
words = word_tokenize(temptext.lower())
```

```python
words
```

```python
myStopWords = list(punctuation) + stopwords.words('english')
```

```python
wordsNoStopWords = [w for w in words if w not in myStopWords]
```

```python
wordsNoStopWords
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(wordsNoStopWords)
```

```python
freq
```

```python
for i in sorted(freq, key=freq.get, reverse=True)[:20]:
    print(i,freq[i])
```

```python
extrastop = ['``',"''","'re","'s","'re",'``',"''","'ll","--","\'\'","...",
             "n\'t",'one','would','use','subject','from',"\'m","\'ve"]
```

```python
myStopWords = list(punctuation) + stopwords.words('english') + extrastop
```

```python
[w for w in word_tokenize(moviereviews[0].lower()) if w not in myStopWords]
```

```python
listOfNewsWords = []
for i in moviereviews:
    listOfNewsWords.append([w for w in word_tokenize(i.lower()) if w not in myStopWords])
```

```python
listOfNewsWords[0]
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
p_stemmer = PorterStemmer()
```

```python
listOfStemmedWords = []
for i in listOfNewsWords:
    listOfStemmedWords.append([p_stemmer.stem(w) for w in i])
```

```python
listOfStemmedWords[0]
```

```python
!pip install gensim
```

```python
from gensim import corpora, models
import gensim
```

```python
dictionary = corpora.Dictionary(listOfStemmedWords)
```

```python
print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in listOfStemmedWords]
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, 
                                           num_topics=2, 
                                           id2word = dictionary, 
                                           passes=20)
```

```python
for i in ldamodel.print_topics(num_topics=2, num_words=20):
    print(i)
```
